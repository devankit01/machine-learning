## Choose the Right Estimator/Algorithm for Your Problem

- Sklearn refers to machine learning models and algorithms as estimators.
- Classification Problem: Predicting a category (e.g., heart disease or not).
- Sometimes you'll see `clf` (short for classifier) used as a classification estimator.
- Regression Problem: Predicting a number (e.g., selling price of a car).
- If you're working on a machine learning problem and not sure which model to use with Sklearn, refer to the Sklearn machine learning map: [Sklearn Machine Learning Map](https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html)

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

## 2.1 Picking a machine learning model for a regression problem
Let's use the California Housing dataset - https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_california_housing.html

```python
# Get California Housing dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing
```

```python
housing_df = pd.DataFrame(housing["data"], columns=housing["feature_names"])
housing_df.head()
```

```python
housing_df["MedHouseVal"] = housing["target"]
housing_df["target"] = housing["target"]
housing_df.drop("MedHouseVal", axis=1)
housing_df
```

## Method 1 : Ridge Method


```python
from sklearn.linear_model import Ridge, BayesianRidge, LassoLars
from sklearn import svm

# random seed
np.random.seed(42)

# create the data 
x = housing_df.drop("target", axis=1)
y = housing_df["target"] 

# split data into train and test dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# initialize model


model = LassoLars()
# model = Ridge()
# model = BayesianRidge() 
"""
LassoLars :  43%
Ridge :  99%
BayesianRidge :  100%
"""
model.fit(x_train, y_train)

# check the score of the model
model.score(x_test, y_test)
```

## Method 2 : Ensemble methods -> RandomForestRegressor
The goal of ensemble methods is to combine the predictions of several base estimators built with a given learning algorithm in order to improve generalizability / robustness over a single estimator.

```python
# import RandomForestRegressor model class deom ensemble
from sklearn.ensemble import RandomForestRegressor

# random seed
np.random.seed(42)

# create the data 
x = housing_df.drop("target", axis=1)
y = housing_df["target"] 

# split data into train and test dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# initialize model
from sklearn.linear_model import Ridge

model = RandomForestRegressor()
model.fit(x_train, y_train)

# check the score of the model
model.score(x_test, y_test)
```

## Picking a machine learning model for a classification problem

```python
 
heart_disease = pd.read_csv("https://raw.githubusercontent.com/mrdbourke/zero-to-mastery-ml/master/data/heart-disease.csv")
heart_disease.head()
```

## Method 1 : LinearSVC estimator

```python
# Import the LinearSVC estimator class
from sklearn.svm import LinearSVC

# Setup random seed
np.random.seed(50) # only , you can see same result

# Make the data
x = heart_disease.drop("target", axis=1)
y = heart_disease["target"]

# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

# Instantiate LinearSVC
model = LinearSVC(max_iter=10000)
model.fit(x_train, y_train)

# Evaluate the LinearSVC
model.score(x_test, y_test)
```

## Method 2 :  Ensemble methods -> RandomForestRegressor 

```python
# Import the RandomForestClassifier estimator class
from sklearn.ensemble import RandomForestClassifier

# Setup random seed
np.random.seed(42)

# Make the data
X = heart_disease.drop("target", axis=1)
y = heart_disease["target"]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Instantiate Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)

# Evaluate the Random Forest Classifier
clf.score(X_test, y_test)
```
