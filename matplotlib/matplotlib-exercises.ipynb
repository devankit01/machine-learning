# Introduction to Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot([1, 2, 3, 4, 5]);
```

```python
x = [1, 3, 4, 6]
y = [10, 50, 30, 40]

# Method 1
plt.plot(x, y)
```

```python
# Method 2
fig, ax = plt.subplots()
ax.plot(x, y)

# customize plot
ax.set(title="Simple Plot", xlabel="x-label", ylabel="y-label")

# save graph
fig.savefig("chart_1.png")
```

```python
# Line  Plot
import numpy as np

x = np.linspace(0, 10, 100)


fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# bar plot 
data = {
    "Backend" : 10, 
    "Frontend" : 20,
    "DevOps" : 5,
    "HR" : 2
}
fig, ax = plt.subplots()
ax.bar(data.keys(), data.values())
ax.set(title="Company Size")
```

```python
# horizontal bar plot
fig, ax = plt.subplots()
ax.barh(list(data.keys()), list(data.values()))
ax.set(title="Company Size")
```

```python
# histograph
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

```python
# subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2 , figsize=(10, 5))

ax1.plot(x, x/2)
ax2.scatter(np.random.random(100), np.random.random(100))
ax3.bar(data.keys(), data.values())
ax4.hist(np.random.randn(1000));
```

## Plotting from Pandas data

```python
# ts : timeseries
ts = pd.Series(np.random.randn(1000), index=pd.date_range("01/01/2021", periods=1000))
ts = ts.cumsum() # cumulative sum
ts.plot()
```

```python
import pandas as pd

car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
# convert price format
car_sales["Price"] = car_sales["Price"].str.replace(r'[\$,\.]', '', regex=True)
car_sales["Price"] = car_sales["Price"].astype("int64")
car_sales["Price"] = car_sales["Price"] / 100
car_sales
```

```python
# add date column
car_sales["Date"] = pd.Series(pd.date_range("01/01/2023", periods=len(car_sales)))

car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
# plot total sales
car_sales.plot(x="Date", y="Total Sales")
```

```python
# Scatter plot for Price/Odometer
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```

```python
# histograph
car_sales["Odometer (KM)"].plot.hist(bins=10);
```

```python
# heart disease dataframe
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# create histograph for age distribution
heart_disease["age"].plot.hist(bins=10);
```

```python
# plot two data with age factor
fig, ax = plt.subplots(figsize = (10, 6))

# plot data
scatter = ax.scatter(x=heart_disease["age"], 
                     y = heart_disease["chol"],
                     c = heart_disease["target"])
ax.set(title="Heart disease and Cholestrol Level", xlabel="Age", ylabel="Cholestrol")

# add legend
ax.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax.axhline(heart_disease["chol"].mean(), linestyle="--")
```
